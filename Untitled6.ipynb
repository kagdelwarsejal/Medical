```python
!pip install pillow opencv-python-headless
import cv2
import numpy as np
import imageio
from PIL import Image
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, MiniBatchKMeans
```

```python
# Convert TIFF to JPG
def convert_tif_to_jpg(tif_path, jpg_path):
    # Read the TIFF image
    image = imageio.v2.imread(tif_path)
    # Normalize to 0-255 and convert to uint8
    image = (255 * (image - np.min(image)) / (np.max(image) - np.min(image))).astype(np.uint8)
    # Convert to RGB if the image has an alpha channel
    if image.ndim == 3 and image.shape[2] == 4:  # RGBA
        image = image[:, :, :3]  # Remove alpha channel
    # Save the image in JPG format
    Image.fromarray(image).save(jpg_path, 'JPEG')
```

```python
# Preprocess the image
def preprocess_image_color(image_path):
    # Load the image in color
    image = cv2.imread(image_path)
    # Normalize each color channel separately
    normalized_image = np.zeros_like(image)
    for i in range(image.shape[2]):  # Iterate over color channels
        normalized_image[:, :, i] = cv2.normalize(image[:, :, i], None, 0, 255, cv2.NORM_MINMAX)
    return normalized_image
```

```python
# Convert RGB to HSV
def rgb_to_hsv(image):
    # Convert the image from RGB to HSV
    return cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
```

```python
# Apply gradient colors based on HSV values
def apply_gradient_colors(hsv_image):
    # Create an output image with gradient colors
    output_image = np.zeros_like(hsv_image)

    # Define the color gradients for the HSV channels
    lower_color = np.array([30, 255, 255])  # Starting color (green) in HSV
    upper_color = np.array([90, 255, 255])  # Ending color (yellow) in HSV

    # Calculate the gradient for the hue channel
    hue_gradient = np.linspace(lower_color[0], upper_color[0], 256)

    # Apply the hue gradient to the output image
    h = hsv_image[:, :, 0]
    s = hsv_image[:, :, 1]
    v = hsv_image[:, :, 2]
    h_new = hue_gradient[h]
    output_image[:, :, 0] = h_new
    output_image[:, :, 1] = s
    output_image[:, :, 2] = v

    # Convert the HSV image back to BGR for display
    return cv2.cvtColor(output_image, cv2.COLOR_HSV2BGR)
```

```python
# Display the image
def display_image(image, title, color_conversion=cv2.COLOR_BGR2RGB):
    plt.figure(figsize=(10, 5))
    plt.imshow(cv2.cvtColor(image, color_conversion))
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
# Paths to your images
tif_path = '/content/drive/MyDrive/TILED IMAGES/tree_orchard_count_new0_1_2.tif'
jpg_path = '/content/drive/MyDrive/TILED IMAGES/tree_orchard_count_new0_1_2.jpg'
```

```python
# Convert TIFF to JPG
convert_tif_to_jpg(tif_path, jpg_path)

# Preprocess the image
preprocessed_image = preprocess_image_color(jpg_path)

# Convert to HSV
hsv_image = rgb_to_hsv(preprocessed_image)

# Apply gradient colors to the HSV image
gradient_image = apply_gradient_colors(hsv_image)

# Display the images
display_image(preprocessed_image, 'Preprocessed Image')
display_image(hsv_image, 'HSV Image', cv2.COLOR_HSV2RGB)
display_image(gradient_image, 'Gradient Colored Image')
```

```python
# Define HSV color ranges for trees and orchards
# You can adjust these ranges based on your specific needs
tree_lower_bound = np.array([30, 40, 40])  # Example lower bound for trees
tree_upper_bound = np.array([90, 255, 255])  # Example upper bound for trees

orchard_lower_bound = np.array([10, 100, 100])  # Example lower bound for orchards
orchard_upper_bound = np.array([50, 255, 255])  # Example upper bound for orchards
```

```python
# Create masks for trees and orchards
tree_mask = cv2.inRange(hsv_image, tree_lower_bound, tree_upper_bound)
orchard_mask = cv2.inRange(hsv_image, orchard_lower_bound, orchard_upper_bound)

# Combine masks to differentiate between trees, orchards, and other elements
combined_mask = cv2.bitwise_or(tree_mask, orchard_mask)
```

```python
# Apply masks to the gradient-colored image
masked_gradient_image = cv2.bitwise_and(gradient_image, gradient_image, mask=combined_mask)
```

```python
# Display the masked gradient image
display_image(masked_gradient_image, 'Masked Gradient Colored Image')
```

```python
# Step 1: Flatten the image to a 2D matrix
reshaped_image = hsv_image.reshape((-1, hsv_image.shape[2]))

# Step 2: Scale the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(reshaped_image)

# Step 3: Apply PCA to reduce dimensions
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)

# Perform K-means clustering
num_clusters = 3  # Example number of clusters
kmeans = KMeans(n_clusters=num_clusters, random_state=0)
clusters = kmeans.fit_predict(pca_data)

# Reshape the clusters to match the original image shape
clustered_image = clusters.reshape(hsv_image.shape[0], hsv_image.shape[1])

# Display the clustered image
plt.figure(figsize=(10, 10))
plt.imshow(clustered_image, cmap='nipy_spectral')
plt.title('Clustered Image')
plt.axis('off')
plt.show()

# Apply mini-batch k-means clustering
minibatch_kmeans = MiniBatchKMeans(n_clusters=num_clusters, random_state=0)
minibatch_kmeans.fit(scaled_data)
clusters = minibatch_kmeans.labels_
clustered_image = clusters.reshape(hsv_image.shape[0], hsv_image.shape[1])

# Display the clustered image
plt.figure(figsize=(10, 10))
plt.imshow(clustered_image, cmap='nipy_spectral')
plt.title('Clustered Image')
plt.axis('off')
plt.show()
```

```python
# Display the gradient image and clustered images side by side
def display_side_by_side(images, titles, color_conversions):
    plt.figure(figsize=(20, 10))
    for i in range(len(images)):
        plt.subplot(1, len(images), i + 1)
        if color_conversions[i] is not None:
            plt.imshow(cv2.cvtColor(images[i], color_conversions[i]))
        else:
            plt.imshow(images[i])
        plt.title(titles[i])
        plt.axis('off')
    plt.show()

# Convert clustered images to uint8 for display
clustered_image_kmeans_uint8 = clustered_image.astype(np.uint8)
clustered_image_minibatch_uint8 = clustered_image.astype(np.uint8)

# Display images
display_side_by_side(
    [gradient_image, clustered_image_kmeans_uint8, clustered_image_minibatch_uint8],
    ['Gradient Image', 'K-means Clustered Image', 'MiniBatch K-means Clustered Image'],
    [cv2.COLOR_BGR2RGB, None, None]
)

```
